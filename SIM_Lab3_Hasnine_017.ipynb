<H2><b>Student's Assignment 1: Basic Matrix Operations</b></H2><br>

<b>Tasks:</b>

1. Create two 3×3 matrices with random integers in Python. <br>
2. Perform addition, subtraction, and multiplication operations on these matrices. <br>
3. Compute the determinant, inverse, and rank of one of the matrices. <br>

```python
# Create two 3x3 random integer matrices
A = np.random.randint(1, 10, (3, 3))
B = np.random.randint(1, 10, (3, 3))

print("Matrix A:\n", A,"\n")
print("Matrix B:\n", B)
```

```python
#addition, subtraction, and multiplication
add = A + B
sub = A - B
mul = np.dot(A, B)

print("\nA + B:\n", add)
print("\nA - B:\n", sub)
print("\nA x B:\n", mul)
```

```python
# Determinant
detA = np.linalg.det(A)
print("\nDeterminant of A:", detA)

# Rank
rankA = np.linalg.matrix_rank(A)
print("Rank of A:", rankA)

# Inverse (if determinant ≠ 0)

if detA != 0:
    inverseA = np.linalg.inv(A)
    print("Inverse of A:\n", inverseA)
else:
    print("Matrix A is singular. Inverse does not exist.")
```

<b>Task 3:</b> 
Matrix Operations and Visualization: <br>
a. Create two 4x4 matrices with random values of your choice. <br>
b. Perform the following operations:<br>
i. Matrix Addition<br>
ii. Matrix Subtraction<br>
iii. Matrix Multiplication (ensure matrices are compatible for multiplication)<br>

c. For each of these operations, visualize the result using bar plots. Each bar should represent one element of the matrix.<br>

```python
# Create two 4x4 matrices with random integer values

X = np.random.randint(1, 11, (4, 4))
Y = np.random.randint(1, 11, (4, 4))

print("Matrix X:\n", X)
print("\nMatrix Y:\n", Y)
```

```python
# Perform matrix operations

add_xy = X + Y
sub_xy = X - Y
mul_xy = np.dot(X, Y)

print("\nX + Y:\n", add_xy)
print("\nX - Y:\n", sub_xy)
print("\nX x Y:\n", mul_xy)
```

```python
# Bar plot for Addition

plt.bar(range(16), add_xy.flatten())

plt.title("Bar Plot of X + Y")
plt.xlabel("Element Index")
plt.ylabel("Value")

plt.show()
```

```python
# Bar plot for Subtraction

plt.bar(range(16), sub_xy.flatten())

plt.title("Bar Plot of X - Y")
plt.xlabel("Element Index")
plt.ylabel("Value")

plt.show()
```

```python
# Bar plot for Multiplication

plt.bar(range(16), mul_xy.flatten())

plt.title("Bar Plot of X x Y")
plt.xlabel("Element Index")
plt.ylabel("Value")

plt.show()
```

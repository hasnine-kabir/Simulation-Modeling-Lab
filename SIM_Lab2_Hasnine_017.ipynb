```python
#personal matrix creation
A = np.array([[9, 2], 
              [14, 3]])

print("Matrix A: \n", A)
```

<b>Task 2:</b>

```python
# Shape
print("Shape of A:", A.shape)

# Determinant
detA = np.linalg.det(A)
print("Determinant of A:", detA)

# Rank
rankA = np.linalg.matrix_rank(A)
print("Rank of A:", rankA)

# Eigenvalues
eigenvalues = np.linalg.eigvals(A)
print("Eigenvalues of A:", eigenvalues)

# Inverse
if detA != 0:
    invA = np.linalg.inv(A)
    print("Inverse of A:\n", invA)
else:
    print("Inverse does not exist because determinant is 0")
```

Value Change Experiment

Creating a new matrix B by changing exactly one value in A:

```python
# Create B by changing one value (A[0,1] = 2 -> 3)
B = np.array([[9, 3],
              [14, 3]])

print("Matrix B (changed A[0,0] from 2 to 3):\n", B)
```

<b>Task 4: </b>

```python
#Mat B operations
# Shape
print("Shape of B:", B.shape)

# Determinant
detB = np.linalg.det(B)
print("Determinant of B:", detB)

# Rank
rankB = np.linalg.matrix_rank(B)
print("Rank of B:", rankB)

# Eigenvalues
eigenvalues_B = np.linalg.eigvals(B)
print("Eigenvalues of B:", eigenvalues_B)

# Inverse
if detB != 0:
    invB = np.linalg.inv(B)
    print("Inverse of B:\n", invB)
else:
    print("Inverse does not exist because determinant is 0")
```
